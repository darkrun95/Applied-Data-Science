```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd
import numpy as np

%matplotlib notebook

remove_col_names = []
fig = plt.figure(figsize=(9.5,5.8))

def my_func(x):
    global remove_col_names
    if x.split("-")[1] == "02" and x.split("-")[2] == "29":
        remove_col_names.append(x)
    return x

def leaflet_plot_stations(binsize, hashid):
    global remove_col_names
    df = pd.read_csv('more_data.csv'.format(binsize))
    new_df = df.copy()
    new_df['Date'].apply(my_func)
    new_df['Data_Value'] = new_df['Data_Value'].apply(lambda x: float(x)/10)
    for date in remove_col_names:
        new_df = new_df[new_df["Date"] != date]
        
    new_df['Year'] = new_df['Date'].apply(lambda x:x.split("-")[0])
    range_df = new_df[(new_df['Year'] >= "2005") & (new_df['Year'] <= "2014")]
    alternate_df = new_df[(new_df['Year'] == "2015")]
    
    range_df = range_df.sort_values(by='Date')
    alternate_df = alternate_df.sort_values(by='Date')
    
    range_df = range_df.groupby(by='Date')
    alternate_df = alternate_df.groupby(by='Date')
    return range_df, alternate_df

range_df, alternate_df = leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
max_temp = range_df['Data_Value'].max()
min_temp = range_df['Data_Value'].min()

plt.plot([x for x in range(0, len(max_temp.index.values))], max_temp.values, color="r", linewidth=0.5)
plt.plot([x for x in range(0, len(min_temp.index.values))], min_temp.values, color="b", linewidth=0.5)

ax = plt.gca()
labels = max_temp.index.values

year = "2005"
xticks = []
for label in zip(labels, range(0,len(max_temp.values))):
    label_text = label[0].split("-")
    if label[1] == 0:
        xticks.append(label_text[0])
    elif label_text[0] != year:
        year = label_text[0]
        xticks.append(year)

ticks = ax.xaxis.get_ticklabels()
tick_label = [tick.get_text() for tick in ticks]
if len(tick_label) < len(xticks):
    tick_label.extend([""] * (len(xticks) - len(tick_label)))
else:
    tick_label = tick_label[:len(xticks)]
plt.xticks([x for x in range(0, len(tick_label))], xticks)
    
plt.ylabel('Temperature in degree C')
plt.title('Temperature variations between 2005-2014', fontsize=10)
plt.fill_between([x for x in range(0, len(max_temp.index.values))], max_temp.values, min_temp.values, color='grey', alpha='0.5')
```
